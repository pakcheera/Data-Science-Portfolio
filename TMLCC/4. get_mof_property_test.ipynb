```python
train=pd.DataFrame(data={"MOFname":file})
```

```python
def add_path(fp):
  path="/content/"
  return path+fp+".xyz"
```

```python
train["path"]=train[["MOFname"]].apply(add_path)
```

```python
def mof_p(path):
  molsys = pw.MolecularSystem.load_file(path)
  mol = molsys.system_to_molecule()
  p=mol.calculate_pore_volume_opt()
  x,y,z=mol.calculate_centre_of_mass()
  return p,x,y,z
```

```python
train["pore_volume"]=""
train["centre_of_mass_x"]=""
train["centre_of_mass_y"]=""
train["centre_of_mass_z"]=""
for i in range(len(train)):
  train["pore_volume"][i],train["centre_of_mass_x"][i],train["centre_of_mass_y"][i],train["centre_of_mass_z"][i]=mof_p(train["path"][i])
  if i%100==0:
        print(i)
```

```python
d=pd.read_csv(path+"test_with_smiles.csv")
```

```python
d=d.drop(columns=["Unnamed: 0"])
```

```python
train=train.drop(columns=["path"])
```

```python
final=pd.merge(train,d,on="MOFname")
```

```python
final=final[['MOFname',  'smiles','pore_volume', 'centre_of_mass_x', 'centre_of_mass_y',
       'centre_of_mass_z', 'volume [A^3]', 'weight [u]',
       'surface_area [m^2/g]', 'void_fraction', 'void_volume [cm^3/g]',
       'functional_groups', 'metal_linker', 'organic_linker1',
       'organic_linker2', 'topology', 'CO2/N2_selectivity',
       'heat_adsorption_CO2_P0.15bar_T298K [kcal/mol]']]
```
