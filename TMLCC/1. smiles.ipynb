```python
import pandas as pd
import numpy as np
import os
```

##trian smiles

```python
data=pd.read_csv(path+"train.csv")
```

```python
data=data[~((data["surface_area [m^2/g]"]==0) | (data["void_fraction"]==0))]
```

```python
def s_cif(x):
  return x.split(".")[0]
```

```python
file=data["MOFname"].values
```

```python
len(file)
```

```python
!unzip "/content/drive/MyDrive/Wanchem colab/Final colab /mof_cif_train.zip"
```

```python
def cif_to_smiles(fname):
    try:
      mol = next(pybel.readfile("cif", fname))
      smi = mol.write(format="smiles")
      return smi.split()[0].strip()
    except:
      return np.nan
```

```python
def add_path(fp):
  path="/content/mof_cif_train/"
  return path+fp+".cif"
```

```python
data["path"]=data[["MOFname"]].apply(add_path)
```

```python
data["smiles"]=""
for i in range(0,len(data)):
    data["smiles"][i]=cif_to_smiles(data["path"][i])
    if i%1000==0:
        data=data.dropna()
        data.to_csv(path+"train_with_smiles.csv")
data=data.dropna()
data.to_csv(path+"train_with_smiles.csv")
```

```python
data=pd.read_csv(path+"train_with_smiles.csv")
data=data.drop(columns="path")
data.to_csv(path+"train_with_smiles.csv")
```
