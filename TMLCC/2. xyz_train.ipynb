```python
data=pd.read_csv(path+"train_with_smiles.csv)
```

```python
file=data["MOFname"].values
```

```python
train=pd.DataFrame(data={"MOFname":file})
```

```python
def add_path(fp):
  path="/content/mof_cif_train/"
  return path+fp+".cif"
```

```python
train["path"]=train[["MOFname"]].apply(add_path)
```

```python
import os 
os.mkdir("/content/mof_xyz_train/")
```

```python
def write_xyz(filename,name):
  mymof = read(filename)
  mymof.write("/content/mof_xyz_train/"+name+".xyz")
```

```python
for i in range(0,len(train)):
      write_xyz(train["path"][i],train["MOFname"][i])
      if i%100==0:
          print(i)
```
